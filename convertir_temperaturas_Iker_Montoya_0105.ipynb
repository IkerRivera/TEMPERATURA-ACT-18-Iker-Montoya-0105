### 1. Fórmulas de Conversión

Para realizar las conversiones entre estas dos escalas de temperatura, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:**
    $$\text{°F} = (\text{°C} \times 1.8) + 32$$
    *(O también expresado como $\text{°F} = (\text{°C} \times \frac{9}{5}) + 32$)*

* **De Fahrenheit a Celsius:**
    $$\text{°C} = \frac{\text{°F} - 32}{1.8}$$
    *(O también expresado como $\text{°C} = (\text{°F} - 32) \times \frac{5}{9}$)*

---

### 2. Ejemplos Resueltos

#### Ejemplo A: Convertir 25°C a °F (Temperatura ambiente cómoda)
1.  **Identificar la fórmula:** $\text{°F} = (\text{°C} \times 1.8) + 32$.
2.  **Sustituir el valor:** $\text{°F} = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** 25°C equivalen a **77°F**.

#### Ejemplo B: Convertir 98.6°F a °C (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $\text{°C} = \frac{\text{°F} - 32}{1.8}$.
2.  **Sustituir el valor:** $\text{°C} = \frac{98.6 - 32}{1.8}$.
3.  **Restar:** $98.6 - 32 = 66.6$.
4.  **Dividir:** $66.6 / 1.8 = 37$.
* **Resultado:** 98.6°F equivalen a **37°C**.

### 3. Código en Python

Este programa es interactivo y utiliza funciones para organizar el código de manera profesional:

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("Ingresa los grados Celsius (°C): "))
    f = (c * 1.8) + 32
    print(f"--- {c}°C equivalen a {f:.2f}°F ---\n")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("Ingresa los grados Fahrenheit (°F): "))
    c = (f - 32) / 1.8
    print(f"--- {f}°F equivalen a {c:.2f}°C ---\n")

def menu():
    """Función principal que controla el flujo del programa"""
    while True:
        print("=== MENÚ DE CONVERSIONES DE TEMPERATURA ===")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")
        
        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.\n")

# Llamada a la función principal para iniciar el programa
if __name__ == "__main__":
    menu()
```

### Explicación del código:
* **`while True`**: Crea un ciclo infinito que permite que el menú se repita hasta que el usuario decida salir.
* **`float()`**: Se usa para que el programa acepte números con decimales (por ejemplo, 36.5).
* **`f"{valor:.2f}"`**: Es un formato para mostrar solo 2 decimales en el resultado final, haciéndolo más legible.
* **`break`**: Rompe el ciclo `while` inmediatamente cuando el usuario elige la opción 3.

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("Ingresa los grados Celsius (°C): "))
    f = (c * 1.8) + 32
    print(f"--- {c}°C equivalen a {f:.2f}°F ---\n")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("Ingresa los grados Fahrenheit (°F): "))
    c = (f - 32) / 1.8
    print(f"--- {f}°F equivalen a {c:.2f}°C ---\n")

def menu():
    """Función principal que controla el flujo del programa"""
    while True:
        print("=== MENÚ DE CONVERSIONES DE TEMPERATURA ===")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("programa realizado por Iker Montoya NL = 39")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.\n")

# Llamada a la función principal para iniciar el programa1
if __name__ == "__main__":
    menu()
````
